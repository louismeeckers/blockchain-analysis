# Create JSON files for mapping

## Define structure of our nodes

```python
class Block:
    def __init__(self, block_hash, block_height, timestamp, prev_block_hash):
        self.height = block_height
        self.hash = block_hash
        self.timestamp = timestamp
        self.time_gmt = datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%dT%H:%M:%S')
        self.prev_block_hash = prev_block_hash
        
class Address:
    def __init__(self, address_hash):
        self.hash = address_hash
        
class Transaction():
    def __init__(self, transaction_hash, tx_index, block_hash, outputs):
        self.hash = transaction_hash
        self.index = tx_index
        self.block_hash = block_hash
        self.outputs = outputs
        
class UTXO():
    def __init__(self, tx_index, n, value, address, input_of_tx_index):
        self.tx_index = tx_index
        self.n = n
        self.value = value
        self.address = address
        if input_of_tx_index is not None:
            self.input_of_tx_index = input_of_tx_index
            self.address = address

class CoinbaseUTXO():
    def __init__(self, block_hash, value, input_of_tx_index):
        self.block_hash = block_hash
        self.value = value
        self.input_of_tx_index = input_of_tx_index
```

## Loop over '*n*' blocks (start with '*hash_block_i*')

```python
blocks = []
addresses = []
transactions = []
utxos = []
coinbase_utxos = []

# block 0: 000000000019d6689c085ae165831e934ff763ae46a2a6c172b3f1b60a8ce26f
# block 1: 00000000839a8e6886ab5951d76f411475428afc90947ee320161bbf18eb6048
# 100th block: 00000000b69bd8e4dc60580117617a466d5c76ada85fb7b87e9baea01f9d9984
# old block: 0000000000000000000060394b3a68de6834b6be90e47b61f34d0b895ff04d48

n = 100
hash_block_i = '0000000000000000000060394b3a68de6834b6be90e47b61f34d0b895ff04d48'

for b in tqdm(range(n)):
    block = getBlock(hash_block_i)
    blocks.append(Block(block['hash'], block['height'], block['time'], block['prev_block']))
    
    for iTx, tx in enumerate(block['tx']):
        outputs = []
        
        if iTx == 0:
            coinbase_utxos.append(CoinbaseUTXO(hash_block_i, 0, tx['tx_index'])) # value of utxo is always 0
        
        for o in tx['out']:
            outputs.append(str(o['tx_index']) + '_' + str(o['n'])) # for transaction outputs relations
            
            if o['spending_outpoints']: # if has been spent in a transaction
                utxos.append(UTXO(o['tx_index'], o['n'], o['value'], o.get('addr'), o['spending_outpoints'][0]['tx_index']))
                addresses.append(Address(o.get('addr')))
            else: # if has not been spent in a transaction (yet)
                utxos.append(UTXO(o['tx_index'], o['n'], o['value'], o.get('addr'), None))

        transactions.append(Transaction(tx['hash'], tx['tx_index'], hash_block_i, outputs))

    hash_block_i = block['next_block'][0]
```

## Create files in data/json/

```python
# JSON Blocks
json_data = json.dumps([block.__dict__ for block in blocks], indent=4, ensure_ascii=False)
text_file = open("data/json/blocks.json", "w", encoding='utf-8')
text_file.write(json_data)
text_file.close()
```

```python
# JSON Addresses
json_data = json.dumps([address.__dict__ for address in addresses], indent=4, ensure_ascii=False)
text_file = open("data/json/addresses.json", "w", encoding='utf-8')
text_file.write(json_data)
text_file.close()
```

## Creation of the graph

```python
g = rdflib.Graph()

blocks = open("data/turtle/blocks.ttl", encoding="ISO-8859-1")
addresses = open("data/turtle/addresses.ttl", encoding="ISO-8859-1")
transactions = open("data/turtle/transactions.ttl", encoding="ISO-8859-1")
utxos = open("data/turtle/utxos.ttl", encoding="ISO-8859-1")
coinbase_utxos = open("data/turtle/coinbase_utxos.ttl", encoding="ISO-8859-1")

g.parse(blocks, format='ttl')
g.parse(addresses, format='ttl')
g.parse(transactions, format='ttl')
g.parse(utxos, format='ttl')
g.parse(coinbase_utxos, format='ttl')
```

## Converting rdflib.Graph to networkx.Graph

```python
mdg = rdflib_to_networkx_multidigraph(g)
dg = rdflib_to_networkx_digraph(g)
# gx = rdflib_to_networkx_graph(g)
```

## Centrality Measures

### Degree Centrality

```python
results = nx.degree_centrality(mdg)
ind_rank = {}

for node_id, rank in sorted(results.items(), key=lambda item: item[1], reverse=True)[:10]:
    print("{:6.3f} {}".format(rank, node_id))
```

### Closeness Centrality

```python
results = nx.degree_centrality(mdg)
ind_rank = {}
topn = 0
for node_id, rank in sorted(results.items(), key=lambda item: item[1], reverse=True):
    if node_id.startswith('https://w3id.org/um/ken4256/ingredient/'):
        print("{:6.5f} {}".format(rank, node_id))
        topn += 1
        if topn > 10:
            break
```
